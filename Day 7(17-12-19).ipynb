```python
#palindrome using functions
def palindrome(n):
    i = 0
    m = n
    while n != 0:
            i = i*10+(n%10)
            n = n//10
    if m == i :
        return True
    else :
        return False
palindrome(121)
```

### Types of arguments
##### 1.Required positional arguments
##### 2.Default arguments
##### 3.key-word argument
##### 4.variable-length argument

```python
#required  positinal
def add(x,y) :
    return x+y
add(10)
```

```python
#default arguments
def add(x,y=50) :#default
    return x+y
add(10)
```

```python
#key word argument
def add(x,y) :
    print(x)
    print(y)
    return x+y
add(y=10,x=20)#key word arguments
```

```python
#variable-length argument
def add(*x):
    s = 0
    for i in x:
            s += i
    return s
print(add(12,23))
print(add(10,20,30,40,50))   
```
